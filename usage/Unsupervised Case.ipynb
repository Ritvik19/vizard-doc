```python
import pandas as pd
import vizard
```

```python
df = pd.read_csv('../input/cinema-tickets.csv')
print(df.shape)
df.head()
```

```python
df['day_of_week'] = pd.to_datetime(df['date']).dt.weekday
```

```python
df.describe()
```

```python
class config:
    PROBLEM_TYPE = 'unsupervised'
    CONTINUOUS_INDEPENDENT_VARIABLES = ['tickets_sold', 'ticket_price', 'capacity']
    CATEGORICAL_INDEPENDENT_VARIABLES = ['month', 'quarter', 'day_of_week']
```

```python
viz = vizard.Vizard(df, config)
```

# Missing Value Analysis

```python
viz.check_missing();
```

```python
viz.count_missing();
```

# Analysis of values in all Columns

```python
viz.count_unique();
```

# Univariate Analysis

```python
viz.categorical_variables();
```

```python
viz.continuous_variables();
```

# Bivariate Analysis

```python
viz.pairwise_scatter();
```

```python
viz.pairwise_violin();
```

```python
viz.pairwise_crosstabs();
```

# Trivariate Analysis

```python
viz.trivariate_bubbleplot('tickets_sold', 'ticket_price', 'capacity');
```

```python
viz.trivariate_scatterplot('tickets_sold', 'ticket_price', 'day_of_week');
```

```python
viz.trivariate_violinplot('day_of_week', 'ticket_price', 'quarter');
```

```python
viz.pair_plot();
```

```python
viz.corr_plot();
```
