```python
import pandas as pd
import vizard
```

```python
df = pd.read_csv('../input/insurance.csv')
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
class config:
    PROBLEM_TYPE = 'regression'
    DEPENDENT_VARIABLE = 'charges'
    CATEGORICAL_INDEPENDENT_VARIABLES = [
        'sex', 'smoker', 'region'
    ]
    CONTINUOUS_INDEPENDENT_VARIABLES = [
        'age', 'bmi', 'children'
    ]
```

```python
viz = vizard.Vizard(df, config)
```

# Missing Value Analysis

```python
viz.check_missing();
```

```python
viz.count_missing();
```

# Analyses of values in all Columns

```python
viz.count_unique();
```

# Target Column Distribution

```python
viz.dependent_variable();
```

# Segmented Univariate Analysis

```python
viz.categorical_variables();
```

```python
viz.continuous_variables();
```

# Bivariate Analysis

```python
viz.pairwise_scatter();
```

```python
viz.pairwise_violin();
```

```python
viz.pairwise_crosstabs();
```

# Trivariate Analysis

```python
viz.trivariate_bubbleplot('age', 'bmi', 'children');
```

```python
viz.trivariate_scatterplot('age', 'bmi', 'smoker');
```

```python
viz.trivariate_violinplot('region', 'age', 'sex');
```
