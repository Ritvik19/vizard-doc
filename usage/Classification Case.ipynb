```python
import pandas as pd
import vizard
```

```python
df = pd.read_csv('../input/titanic-train.csv')
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
class config:
    PROBLEM_TYPE = 'classification'
    DEPENDENT_VARIABLE = 'Survived'
    CATEGORICAL_INDEPENDENT_VARIABLES = [
        'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked'
    ]
    CONTINUOUS_INDEPENDENT_VARIABLES = [
        'Age', 'Fare'
    ]
    TEXT_VARIABLES = ['Name', 'Cabin', 'Ticket']
```

```python
viz = vizard.Vizard(df, config)
```

# Missing Value Analysis

```python
viz.check_missing();
```

```python
viz.count_missing();
```

```python
viz.count_missing_by_group('Survived');
```

# Analyses of values in all Columns

```python
viz.count_unique();
```

```python
viz.count_unique_by_group('Survived');
```

# Target Column Distribution

```python
viz.dependent_variable();
```
